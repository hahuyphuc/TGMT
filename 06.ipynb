```python
# Import lớp Image
from PIL import Image
from imgtools import *
from IPython.display import display
```

```python
# Đường dẫn ảnh
path = 'C:/Users/huyph/Documents/TGMT/IMG/06.jpg'
```

```python
img = load_image(path)
display(img)
```

```python
# Chuyển từ ảnh RGB sang ảnh xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
