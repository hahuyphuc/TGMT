```python
# Import lớp Image
from PIL import Image
from imgtools import *
from IPython.display import display
```

```python
# Đường dẫn ảnh
path = 'C:/Users/huyph/Documents/TGMT/IMG/04.jpg'
```

```python
# Đọc ảnh 
img = load_image(path)
```

```python
display(img)
```

```python
# Kích thước ban đầu:
print(img.size)
```

```python
# Định vị khu vực cắt
region = (50, 75, 140, 180)
```

```python
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (10, 10)
img.paste(cropped_image, paste_position)
display(img)
```
